## *변수의 타입

int => 숫자

float => 소수

str => 작은 따옴표(주로 쓰임) or 큰 따옴표

ex. str의 n번째 꺼 지목하기(처음은 0)

list => 대괄호 속에 여러 개가 들어있음

list 안에 어떤 타입이든 들어와도 됨(int이든 str이든 list이든 상관 x)

ex. list의 n 번째 꺼 지목하기

    a = [1. '2', [3, '4']]    a[2] = [3,'4']

dict => 순서가 없음 / a,b와 같은 index로 표현

[ ].join(tokens) => 리스트의 개별 단어들에 공백을 넣어줌
